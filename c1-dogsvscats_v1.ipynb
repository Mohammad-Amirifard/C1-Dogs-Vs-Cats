```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import tensorflow as tf
from tensorflow import keras
import os
from keras import layers
from keras import models
from keras.layers import Dropout
from keras import optimizers
from sklearn.model_selection  import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import callbacks
```

<p id="part4"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">4-Create a dataframe from train data
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
categories = []
for filename in filenames:
    category= filename.split('.')[0]
    if category == 'cat':
        categories.append('0')
    else:
        categories.append('1')

df = pd.DataFrame({'filename':filenames,'category':categories})
print('df.shape:',df.shape)
df.head()
```

```python
print('Number of dogs in the train dataset:',df.category[df.category=='1'].count())
print('Number of cats in the train dataset:',df.category[df.category=='0'].count())
df['category'].value_counts().plot.bar(color=['green','blue'])
```

<p id="part5"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5- Create Model</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
Image_Size = (256,256)
Channel_Size = 3
Batch_Size = 64

# Use CNN Artitechture
def create_model():
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu',
                            input_shape=(256,256,3)))
    model.add(layers.MaxPooling2D((2, 2)))
    
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    
    model.add(layers.Conv2D(256, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(128,activation='relu'))
    model.add(Dropout(0.2))
    
    model.add(layers.Dense(256,activation='relu'))
    model.add(Dropout(0.5))
    
    model.add(layers.Dense(2, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'],)
    return model

my_model = create_model()
```

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5-1 Model Summary</p>

```python
my_model.summary()
```

<p id="part7"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5-2 Split data to train and test section</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
train_df, validation_df = train_test_split(df, test_size=0.2, random_state=45)
train_df = train_df.reset_index(drop=True)
validation_df = validation_df.reset_index(drop=True)
print(f'train_df.shpae:{train_df.shape} & validation_df.shape:{validation_df.shape}')
```

<p id="part8"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5-3 Genereate data by ImageDataGenereator</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
# We generate data and do data agumentation to avoid overfitting

# Train_Section:
train_datagen = ImageDataGenerator(
    rotation_range=15,
    rescale=1./255,
    shear_range=0.1,
    zoom_range=0.4,
    horizontal_flip=True)

train_generator = train_datagen.flow_from_dataframe(
    train_df,
    './train',
    x_col='filename',
    y_col='category',
    target_size=Image_Size,
    class_mode='categorical',
    batch_size=Batch_Size
)


#Validation_Section:---> Note: You don not need to do data agumentation on validation data
validation_datagen = ImageDataGenerator(
    rescale=1./255)

validation_generator = train_datagen.flow_from_dataframe(
    validation_df,
    './train',
    x_col='filename',
    y_col='category',
    target_size=Image_Size,
    class_mode='categorical',
    batch_size=Batch_Size
)
```

<p id="part9"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5-4 Use Callbacks to have better contorl on training</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
# 1st_callbak : EarlyStoping
My_EarlyStopping_callback = callbacks.EarlyStopping(monitor="val_loss",
                                                 patience=3,
                                                 verbose=1,
                                                 mode="auto")


#2nd_callback : ModelCheckPoint
checkpoint_path = './Checkpint_Folder/cp.ckpt' # The path to save wights of the model after training
checkpoint_dir =os.path.dirname(checkpoint_path)
My_ModelCheckpoint_callback =callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                          metrics='val_loss',
                                          verbose=1,
                                          save_weights_only = True,
                                          save_best_only=True,
                                          mode='auto',
                                          save_freq ='epoch')

My_callbacks = [My_EarlyStopping_callback, My_ModelCheckpoint_callback]
```

<p id="part10"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">5-5 Fit model</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
History = my_model.fit(
    train_generator,
    epochs=15,
    validation_data=(validation_generator),
    batch_size = Batch_Size,
    callbacks = My_callbacks
)
```

<p id="part11"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">6- Plot results</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
accuracy = History.history['accuracy']
val_accuracy = History.history['val_accuracy']
loss = History.history['loss']
val_loss = History.history['val_loss']

epochs = range(len(accuracy))

plt.plot(epochs,accuracy, label='accuracy',color='red')
plt.plot(epochs,val_accuracy, label='val_accuracy',color='blue')
plt.title('Training and validation loss')
plt.legend()
plt.figure()

plt.plot(epochs,loss,label='loss',color='black')
plt.plot(epochs,val_loss,label='val_loss',color='green')
plt.title('Training and validation loss')
plt.legend()
plt.figure()
```

<p id="part12"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">7- Evaluation on validation data with best model</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
best_model = create_model()
best_model.load_weights(checkpoint_path)
loss,accuracy = best_model.evaluate(validation_generator)
print('loss on validation data:',loss*100)
print('accuracy on validation data', accuracy*100)
```

<p id="part13"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">8- Create a dataframe from test data and prepare it</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">


```python
test_filenames = os.listdir('./test1/')
test_df = pd.DataFrame({'filename':test_filenames})
print('test_df.shape:',test_df.shape)
test_df.head()
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_dataframe(
    test_df, 
    "./test1/", 
    x_col='filename',
    y_col=None,
    class_mode=None,
    target_size=Image_Size,
    batch_size=Batch_Size,
    shuffle=False
)
```

<p id="part14"></p>

<p style="font-family: Times New Roman; font-size: 20px; font-style: bold; font-weight: bold; letter-spacing: 0px; color: #0000FF">9- Prediction on test data</p>
<hr style="height: 1px; border: 1; background-color: #0000FF">

```python
Predict = best_model.predict(test_generator)
# add the results to test_df as category column
test_df['category'] = np.argmax(Predict, axis=-1)
```
